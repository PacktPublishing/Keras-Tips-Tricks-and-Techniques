# AutoKeras
This is a notebook that demonstrates how to use autokeras. First make sure to install autokeras. Check out [http://autokeras.com](http://autokeras.com)

```python
import autokeras as ak
from keras.datasets import mnist
```

First, let's import MNIST to show the AutoKeras image classifier

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train = x_train.reshape(x_train.shape + (1,))
x_test = x_test.reshape(x_test.shape + (1,))
```

```python
# Search and train the classifier.
# By default it will use 1000 epochs unless specified.
clf = ak.ImageClassifier(max_trials=10, seed=None)
clf.fit(x_train, y_train,validation_split=0.2, epochs=10)
y = clf.predict(x_test, y_test)
```

Let's import cifar10

```python
from sklearn.metrics import classification_report
from keras.datasets import cifar10
```

```python
((trainX, trainY), (testX, testY)) = cifar10.load_data()
trainX = trainX.astype("float") / 255.0
testX = testX.astype("float") / 255.0

labelNames = ["airplane", "automobile", "bird", "cat", "deer",    "dog", "frog", "horse", "ship", "truck"]
```

Let's use the ImageClassifier again from AutoKeras to show how simple it becomes to use

```python
# train our Auto-Keras model
model = ak.ImageClassifier(max_trials=10, seed=None)
model.fit(trainX, trainY, epochs=10, validation_split=0.2)
model.final_fit(trainX, trainY, testX, testY, retrain=True)

# evaluate the Auto-Keras model
score = model.evaluate(testX, testY)
predictions = model.predict(testX)
report = classification_report(testY, predictions,  target_names=labelNames)
```

You can also use AutoKeras for Text based scenarios. Let's load the IMDB review dataset to demonstrate

```python
from keras.datasets import imdb
import numpy as np
```

```python
max_features = 20000
index_offset = 3  # word index offset

(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=max_features,index_from=index_offset)
x_train = x_train
y_train = y_train.reshape(-1, 1)
x_test = x_test
y_test = y_test.reshape(-1, 1)

word_to_id = imdb.get_word_index()
word_to_id = {k: (v + index_offset) for k, v in word_to_id.items()}
word_to_id["<PAD>"] = 0
word_to_id["<START>"] = 1
word_to_id["<UNK>"] = 2

id_to_word = {value: key for key, value in word_to_id.items()}
x_train = list(map(lambda sentence: ' '.join(
    id_to_word[i] for i in sentence), x_train))
x_test = list(map(lambda sentence: ' '.join(
    id_to_word[i] for i in sentence), x_test))
x_train = np.array(x_train, dtype=np.str)
x_test = np.array(x_test, dtype=np.str)
```

```python
clf = ak.TextClassifier(max_trials=3, seed=5)
clf.fit(x_train, y_train, validation_split=0.2)
```

You can also use AutoKeras for regular structured data as well.

```python
clf = ak.StructuredDataClassifier(seed=5, max_trials=3)
clf.fit(x='train.csv', y='survived', validation_split=0.2)
clf.evaluate(x='eval.csv', y='survived')
```

Finally, you can also use AutoKeras for more complex scenarios such as AutoML on a FunctionalAPI based model

```python
from keras import preprocessing
from keras_preprocessing import sequence

max_features = 20000
max_words = 400
(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=max_features,index_from=3)
x_train = sequence.pad_sequences( x_train, maxlen=max_words)
x_test = sequence.pad_sequences(x_test, maxlen=max_words)
print(x_train.dtype)
print(x_train[:10])
input_node = ak.Input()
output_node = input_node
output_node = ak.EmbeddingBlock(max_features=max_features)(output_node)
output_node = ak.ConvBlock()(output_node)
output_node = ak.SpatialReduction()(output_node)
output_node = ak.DenseBlock()(output_node)
output_node = ak.ClassificationHead()(output_node)
clf = ak.GraphAutoModel(input_node, output_node, seed=5, max_trials=3)
clf.fit(x_train, y_train, epochs = 5, validation_split=0.2)
```
